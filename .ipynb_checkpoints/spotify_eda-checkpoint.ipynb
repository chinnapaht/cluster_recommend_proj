```python
from spot_clean import clean
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
sns.set()
plt.style.use('classic')
%matplotlib inline
```

```python
df = pd.read_excel("Spotify.xlsx")
```

```python
df = clean(df)
```

## EDA

```python
df.info()
```

```python
df.describe()
```

```python
df_nocat = df.drop(["title","artist","genre","year","add_year","add_month"],axis = 1)
```

```python
plt.figure(figsize=(15, 5))
sns.boxplot(x="variable", y="value", data= pd.melt(df_nocat))
```

From the boxplot we will see that 
- the range of some feature is very strange
- there are some data that have popularity equal 0  that we can identifie as outlier

Plot histogram to look at the specific feature

```python
plotcol = ["dnce" ,"dB", "live" , "dur","spch" , "pop"]
for i in plotcol :
    df[i].hist()
    plt.xlabel(i)
    plt.show()
```

From the histogram we will see that
- some value of the feature is very vary from the mean but we will use it because it is true attribute of the song except popularity 
- every feature seem to be not normal so we have to tranfrom data later if we want to do regression analysis
- some feature distribution is not normal so if we want to do regression analysis we have to transform it to be normal to meet the assumption of regression analysis


```python
sns.pairplot(df_nocat)
```

From the scattermatrix we will see that data doesn't seem to have linear properties and from the histogram we will see that most of the feature are not normal then we might have to use other model that is not linear regression.

```python
plt.figure(figsize=(13, 5))
sns.barplot(x = df.genre.value_counts().index , y = df.genre.value_counts())
plt.xticks(rotation = 90)
plt.show()
```

The release year of the song

```python
plt.figure(figsize=(16, 10))
sns.barplot(x = df.year.value_counts().index , y = df.year.value_counts())
plt.xticks(rotation = 90)
plt.show()
```

I think the year and month of song have be added to spotify might affect the popularity.

Let see this problem occur or not 

```python
pd.pivot_table(df, index = ['add_year','add_month'], values = 'pop')
```

From the table above we will see that we don't have the problem like we assume

Let see the plot between the year song release and popularity

```python
plt.figure(figsize=(16, 10))
sns.barplot(x = df.groupby("year").mean()["pop"].index , y = df.groupby("year").mean()["pop"])
plt.xticks(rotation = 90)
plt.show()
```

Plot the audio characteristics of the feature over the year

```python
plt.figure(figsize=(16, 10))
columns = ["acous","dnce","nrgy","spch","live","val"]
for col in columns:
    x = df.groupby("year")[col].mean()
    ax = sns.lineplot(x=x.index,y=x,label=col)
ax.set_title('Audio characteristics over year')
ax.set_ylabel('Measure')
ax.set_xlabel('Year')
plt.xticks(rotation = 90)
plt.show()
```

From the line graph we will see that
- live and nrgy seem to have an upward trend
- dnce acou val seem to have a downward trend


note that we have a few data from 1983-2003

```python
corr = df.corr()
plt.figure(figsize=(16, 10))
sns.heatmap(corr, cmap="vlag",square = True , annot = True,linewidths = 0.3)
plt.show()
```

From this heatmap we will see that
- the nrgy and acou have high correlation that mean it not meet the assumption of linear regression
like we see before. If we want to predict population base on our feature we might want to use other regression technique more than linear regression.
