```python
from spot_clean import clean
import pandas as pd
import seaborn as sns
import altair as alt
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import KElbowVisualizer
from yellowbrick.cluster import SilhouetteVisualizer
import pickle
import numpy as np
```

```python
df = pd.read_excel("Spotify.xlsx")
df = clean(df)
df_full = df.copy()
df.drop(["title","artist","add_month","add_year" , "pop","year","genre"],axis = 1 , inplace = True)
df
```

## Feature Enginering

```python
df_clus = df.copy()
scaler = StandardScaler()

df_clus[list(df.columns)] = scaler.fit_transform(df_clus)
```

```python
df_clus
```

Reduce dimension of the data by applying PCA

```python
pca = PCA(n_components = 2) 
X_principal = pca.fit_transform(df_clus) 
X_principal = pd.DataFrame(X_principal) 
X_principal.columns = ['P1', 'P2'] 
  
X_principal.head()
```

# Compare the distortion score from the elbow method


```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(4,30))
visualizer.fit(df_clus)       
visualizer.show()
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(4,30))
visualizer.fit(X_principal)        
visualizer.show()
```

From the model that we using three type of data we will see that the data from PCA have lowest distortion score followed by original data and data from TSNE

https://dzone.com/articles/kmeans-silhouette-score-explained-with-python-exam

```python
model = KMeans(13, random_state=42)
visualizer = SilhouetteVisualizer(model, colors='yellowbrick')
visualizer.fit(df_clus)       
visualizer.show()
```

```python
model = KMeans(11, random_state=42)
visualizer = SilhouetteVisualizer(model, colors='yellowbrick')
visualizer.fit(X_principal)       
visualizer.show()
```

In contrast, we will see that the average silhouette score from the TSNE much higher compare to the PCA

```python
kmeans = KMeans(n_clusters=11)
kmeans.fit(X_principal)
```

```python
X_principal["clus"] = kmeans.labels_
X_principal["clus"] = X_principal["clus"].astype('category')
```

```python
sns.scatterplot(data=X_principal, x= "P1", y= "P2", hue="clus")
```

evaluate clustering model https://www.datanovia.com/en/lessons/cluster-validation-statistics-must-know-methods/#silhouette-coefficient

From the 2 scatterplot we can see that the cluster from PCA stay very close to each other compore to the TSNE


I decide to use the data from TSNE to cluster the model cause it have the best silhouette score and we see that our data don't have much linear properties then using PCA may not be appropriate way.

```python
df["clus"] = kmeans.labels_
df_full["clus"] = kmeans.labels_

df["clus"] = df["clus"].astype("category")
```

```python
df_corr = df.copy()
df_corr = pd.get_dummies(df_corr,prefix=["encode"],columns=["clus"])
df.head()
```

```python
corr = df_corr.corr()
plt.figure(figsize=(20, 15))
sns.heatmap(corr, cmap="vlag",square = True , annot = True,linewidths = 0.3)
plt.show()
```

From the heatmap we can see the attribute of each song in each cluster 
 - cluster 1 have a high liveness, a little bit of  energy , loundnesss and low acoustic
 - cluster 2 have a high acoustic , very low energy and loudness
 - cluster 3 have low bpm and a little bit of acoustic
 - cluster 4 have a high valence , energy and little bit of loundness
 - cluster 5 have high liveness and very low bpm
 - cluster 6 have high bpm , energy , loundness ,vey low danceability and acoustic
 - cluster 7 have high spechiness , bpm , valence and low duration 
 - cluster 8 have low valence and acoustic
 - cluster 9 have a liitle bit of acoustic , low valence and energy
 - clustre 10 have high danceability and valence
 - cluster 11 have a little bit of acoustic and high bpm

```python
df_chart = df_full[["genre","clus"]]

alt.Chart(df_chart).mark_bar().encode(
    x='genre:N',
    y='count(genre):Q',
    color='genre:N',
    column='clus:N'
)
```

We will see that each group of the cluster have the same amount of count except for the cluster 8 (cluster 9) that have many Unknow genre

```python
plt.figure(figsize=(16, 10))
ax = sns.barplot(x = df.clus.value_counts().index , y = df.clus.value_counts())
plt.xticks(rotation = 90)
ax.set(xlabel='cluster', ylabel='count of cluster')
plt.show()
```
